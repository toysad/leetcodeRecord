**-217- Contains Duplicate  (Easy)**

Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.
 

Example 1:

- Input: nums = [1,2,3,1]   
- Output: true 


Example 2:   

- Input: nums = [1,2,3,4]   
- Output: false 


Example 3:   

- Input: nums = [1,1,1,3,3,4,3,2,4,2]   
- Output: true
 

Constraints:

- 1 <= nums.length <= 10^5   

- -10^9 <= nums[i] <= 10^9

```python
#s1: HASH TABLE

class Solution(object):
    def containsDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        dic = {}
        for i in nums:
            if dic.__contains__(i):      #to judge if the item is in dictionary keys
                return True
            else:
                dic[i] = 1
        
        return False
```

```python
#s2: Set & List in Python

class Solution(object):
    def containsDuplicate(self, nums):    
        return True if len(set(nums)) < len(nums) else False
```

**-53- Maximum Subarray (Easy)**

Given an integer array nums, find the contiguous subarray (containing at least one number) which has the largest sum and return its sum.

A subarray is a contiguous part of an array.

 

Example 1:

- Input: nums = [-2,1,-3,4,-1,2,1,-5,4]    
- Output: 6   
- Explanation: [4,-1,2,1] has the largest sum = 6.   


Example 2:  

- Input: nums = [1]   
- Output: 1   


Example 3:   

- Input: nums = [5,4,-1,7,8]   
- Output: 23  
 

Constraints:  

- 1 <= nums.length <= 10^5    
- -10^4 <= nums[i] <= 10^4    
 

Follow up: If you have figured out the O(n) solution, try coding another solution using the divide and conquer approach, which is more subtle.

```python
#s1: Greedy    
# 时间复杂度 O(n)， 空间 O(1)
# 核心思想: 若当前指针所指元素之前的数列和小于0，则丢弃当前指针之前的全部数列
class Solution(object):
    def maxSubArray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        rs = -float("inf")
        tmp = 0
        
        #考虑当前数
        for i in nums:
            tmp = tmp + i
            if tmp >= rs:
                rs = tmp
            if tmp < 0:
                tmp = 0
                
        return rs
    
    
#改善代码 （用指针的思维）
class Solution(object):
    def maxSubArray(self, nums):
        
        #初始化的时候，可以把初始和，最大和设定成数组第一个元素
        cur_sum = max_sum = nums[0]
        
        #考虑当前指针之前
        for i in range(0,len(nums)):
            cur_sum = max(nums[i],nums[i]+cur_sum)
            max_sum = max(max_sum,cur_sum)              #比较大小时，利用max()函数，代替不简洁的条件判断语句
            
        return max_sum
```

```python
#S2: DP
# 时间复杂度 O(n)， 空间 O(1)
# 思想： 从左往右看，每个数 是否要和它之前的数构成子序列，条件是 该数与前一个数相加之和 是否比该数本身大 （即前一个数是否大于0）
# 如果比本身大，更新这个数，使其变成与前面一个数的和；否则该数不变
# 遍历后得到的新数列的最大值，即为最大子序列和

class Solution(object):
    def maxSubArray(self, nums):
        for i in range(1,len(nums)):
            if nums[i-1] > 0:
                nums[i] = nums[i]+nums[i-1]
                
        return max(nums)                  
```

对于一个区间$[l,r]$，将它分成 $[l,m]$ 和 $[m+1,r]$ 两个子区间并分治求解，$m=\lfloor\frac{l+r}{2}\rfloor$。   
递归：  
get(a, l, r) 查询a序列[l,r]区间内的最大子段和，**递归**至**区间长度为1**的时候，    
合并问题：    
**如何通过[l,m]区间的信息和[m+1,r]区间的信息合并成区间[l,r]的信息？**
- 维护区间的4个信息：   
        -- lSum 表示[l,r]内以l为左端点的最大子段和   
        -- rSum 表示[l,r]内以r为右端点的最大子段和   
        -- mSum 表示[l,r]内的最大子段和  
        -- iSum 表示[l,r]的区间和  
    
- 合并区间信息  
       -- lSum(l,r) = max(lSum(l,m),(lSum(l,m)+rSum(m+1,r)))
       -- rSum(l,r) = max(rSum(m+1,r),(rSum(m+1,r)+lSum(l,m))
       -- nSum(l,r)有两种可能，跨越两个子区间: =(lSum(m+1,r)+rSum(l,m)) 或 不跨越两个子区间: =nSum(l,m)或=nSum(m+1,r)，三者取最大即可
       -- iSum(l,r) = iSum(l,m)+iSum(m+1,r)    

分析：在此题中，与动态规划的方法相比，分治时间复杂度相同，但使用了递归，并且维护了四个信息的结构体，运行的时间略长，空间复杂度也不如方法一优秀，而且难以理解。但要考虑如下优势  

优势：   
- 不仅可以解决区间[0, n-1]，还可以用于解决任意的子区间[l,r]的问题。  
- 如果我们把[0, n-1]分治下去出现的所有子区间的信息都用堆式存储的方式记忆化下来，即建成一颗真正的树之后，我们就可以在O(logn)的时间内求到任意区间内的答案，我们甚至可以修改序列中的值，做一些简单的维护，之后仍然可以在O(logn)的时间内求到任意区间内的答案。对于大规模查询的情况下，这种方法的优势便体现了出来。   
- 这棵树就是上文提及的一种神奇的数据结构——线段树。这个分治方法类似于「线段树求解最长公共上升子序列问题」的 pushUp 操作。推荐阅读线段树区间合并法解决多次询问的「区间最长连续上升序列问题」和「区间最大子段和问题」  


```python
#s3 代码1
class Solution:
    def maxSubArray(self, nums): 
        n = len(nums)
        #递归终止条件
        if n == 1:
            return nums[0]
        else:
            #递归计算左半边最大子序和
            max_left = self.maxSubArray(nums[0:len(nums) // 2])
            #递归计算右半边最大子序和
            max_right = self.maxSubArray(nums[len(nums) // 2:len(nums)])
        
        #计算中间的最大子序和，从右到左计算左边的最大子序和，从左到右计算右边的最大子序和，再相加
        max_l = nums[len(nums) // 2 - 1]     
        tmp = 0
        for i in range(len(nums) // 2 - 1, -1, -1):
            tmp += nums[i]
            max_l = max(tmp, max_l)
        
        max_r = nums[len(nums) // 2]
        tmp = 0
        for i in range(len(nums) // 2, len(nums)):
            tmp += nums[i]
            max_r = max(tmp, max_r)
        
        #返回三个中的最大值
        return max(max_right,max_left,max_l+max_r)
```

```python
#s3 代码2
# 分治法: 线段树
class Solution :
    def maxSubAray(self, nums):
        class wtevTree : # 线段树
            lSum = 0 # 以左区间为端点的最大子段和
            rSum = 0 # 以右区间为端点的最大子段和
            iSum = 0 # 区间所有数的和
            mSum = 0 # 该区间的最大子段和
            def __init__(self,l, r, i, m) : # 构造函数
                self.lSum = l
                self.rSum = r
                self.iSum = i
                self.mSum = m        


        # 通过既有的属性,计算上一层的属性,一步步往上返回,获得线段树
        def pushUp(self, leftT: wtevTree, rightT: wtevTree) -> wtevTree:
            # 新子段的lSum等于左区间的lSum或者左区间的 区间和 加上右区间的lSum
            l = max(leftT.lSum, leftT.iSum + rightT.lSum)
            # 新子段的rSum等于右区间的rSum或者右区间的 区间和 加上左区间的rSum
            r = max(leftT.rSum + rightT.iSum, rightT.rSum)
            # 新子段的区间和等于左右区间的区间和之和
            i = leftT.iSum + rightT.iSum
            # 新子段的最大子段和,其子段有可能穿过左右区间,或左区间,或右区间
            m = max(leftT.rSum + rightT.lSum, max(leftT.mSum, rightT.mSum))
            return wtevTree(l, r, i, m)
        
        # 递归建立和获得输入区间所有子段的结构
        def getInfo(self, nums: List[int], left: int, right: int) -> wtevTree:
            if (left == right): # 若区间长度为1,其四个子段均为其值
                return wtevTree(nums[left], nums[left], nums[left], nums[left])
            mid = (left + right) >> 1 # 获得中间点mid,右移一位相当于除以2,运算更快
            leftT = getInfo(self, nums, left, mid)
            rightT = getInfo(self, nums, mid + 1, right) # mid+1,左右区间没有交集。
            return pushUp(self, leftT, rightT) # 递归结束后,做最后一次合并
    
        if ( not nums or len(nums) <= 0): # 输入校验
            return 0
        lens = len(nums) # 获取输入长度
        return getInfo(self, nums, 0, lens - 1).mSum
```

**笔记**   

动态规划与分治法(divide and conquer) 有共通和不同：

最优子结构： 如果一个问题的最优解当中包含了子问题的最优解，则该问题具有最优子结构

重叠子问题：用来求解原问题的递归算法反复地解同样的子问题，而不是总是在产生新的子问题。对两个子问题来说，如果它们确实是相同的子问题，只是作为不同问题的子问题出现的话，则它们是重叠的。

动态规划的子问题独立且重叠。
分治法的子问题独立不重叠。

----------------------
**分治法**的思想：  

把一个复杂的问题分成两个或更多的相同或相似的子问题，再把子问题分成更小的子问题……直到最后子问题可以简单的直接求解，原问题的解即子问题的解的合并。

时间复杂度分析：

在解决规模为n的问题时，总是先递归地求解a个规模为$\frac{n}{b}$的子问题，然后在$O(n^d)$时间内将子问题的解合并起来。  
分治算法的运行时间可以通过公式：$T(n) = a T(\lceil\frac{n}{b}\rceil) + O(n^d)$  
则   
$$\begin{equation}
T(n) = \left\{
\begin{array}
O(n^d) & d > \log_b a\\
O(n^d\log_2 n) & d = \log_b a\\
O(n^{\log_b a}) & d < \log_b a
\end{array}
\right.
\end{equation}$$

流程：
1. 划分问题：整个问题划分成多个无关联的子问题。
2. 递归求解：递归调用求解各个子问题。
3. 合并问题：合并子问题的解，形成原始问题的解。
